### Advance operation with Business examples

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id,2024,2025,2026,2027]

sales_data = np.array([
    [1, 1000, 1100, 1200, 1300], #paradise biryani
    [2, 1500, 1600, 1700, 1800], #tasty tacos
    [3, 2000, 2100, 2200, 2300], #delicious pizza
    [4, 2500, 2600, 2700, 2800], #yum burgers
    [5, 3000, 3100, 3200, 3300]  #scrumptious sushi
])

print(" ==== Zomato Sales Data ==== ")
print("\n Sales data shape: ", sales_data.shape)
print("\n 4th row of sales data: \n", sales_data[3])
print("\n First 3 rows of sales data: \n", sales_data[:3])
print("\n First 3 columns of sales data: \n", sales_data[:, :3])
print("\n Sales data for 2025: \n", sales_data[:, 1:])
```

```python
# Total sales per year
total_sales_per_year = np.sum(sales_data[:, 1:], axis=0)
print("\n Total sales per year: ", total_sales_per_year)

# Minimum sales per restaurant
min_sales_per_restaurant = np.min(sales_data[:, 1:], axis=1)    
print("\n Minimum sales per restaurant: ", min_sales_per_restaurant)

# Maximum sales per year
max_sales_per_year = np.max(sales_data[:, 1:], axis=0)
print("\n Maximum sales per year: ", max_sales_per_year)

# Average sales per restaurant
average_sales_per_restaurant = np.mean(sales_data[:, 1:], axis=1)
print("\n Average sales per restaurant: ", average_sales_per_restaurant)
```

```python
cumsum = np.cumsum(sales_data[:, 1:], axis=1)
print("\n Cumulative sales per year: ", cumsum)

plt.figure(figsize=(10, 6))
plt.plot(np.mean(cumsum, axis=0), marker='o')
plt.title('Average Cumulative Sales Over Years')
plt.xlabel('Year')
plt.ylabel('Average Cumulative Sales')
plt.xticks(ticks=range(4), labels=['2024', '2025', '2026', '2027'])
plt.grid()
plt.show()
```

```python
vector1 = np.array([1, 2, 3, 4, 5])
vector2 = np.array([6, 7, 8, 9, 10])
dot_product = np.dot(vector1, vector2)
print("\n Dot product of vector1 and vector2: ", dot_product)

print("Vector Multiplication: ", vector1 * vector2)

print("Vector Addition: ", vector1 + vector2)

angle_radians = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
angle_degrees = np.degrees(angle_radians)
print("\n Angle between vector1 and vector2 (in degrees): ", angle_degrees)
```

```python
restaurant_types = np.array(['Biryani', 'Tacos', 'Pizza', 'Burgers', 'Sushi'])
vectorized_sales = np.vectorize(str.upper)
print("\n Restaurant types in uppercase: ", vectorized_sales(restaurant_types))
```
