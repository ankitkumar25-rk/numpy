### NUMPY FOUNDATION

```python
import numpy as np
```

## Creating array from list


```python
arr_id = np.array([1, 2, 3, 4, 5])
print(arr_id)

arr_id2 = np.array([[1, 2, 3], [4, 5, 6]])
print(arr_id2)
```

### List vs numpy array


```python
py_list = [1, 2, 3, 4, 5]
print("python list multiplication:", py_list * 2)

np_arr = np.array([1, 2, 3, 4, 5])
print("numpy array multiplication:", np_arr * 2)

import time
start = time.time()
py_list = [i*2 for i in range(1000000)]
print("Python list comprehension time:", time.time() - start)

start = time.time()
np_arr = np.arange(1000000) * 2
print("Numpy array multiplication time:", time.time() - start)
```

### creating array from scratch


```python
zeros = np.zeros((3,4))
print("zeroes array: \n", zeros)

ones = np.ones((2,3))
print("ones array: \n", ones)

full = np.full((2,2), 7)
print("full array: \n", full)

random_arr = np.random.random((2,3))
print("random array: \n", random_arr) 

sequence = np.arange(0, 11, 2)
print("sequence array: \n", sequence)
```

### Vector, Matrix and Tensor

```python
vector = np.array([1, 2, 3])
print("vector: \n", vector)

matrix = np.array([[1, 2, 3], [4, 5, 6]])
print("matrix: \n", matrix)

tensor = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
print("tensor: \n", tensor)

```

### Array properties


```python
arr = np.array([[1,2,3],
                [4,5,6]])

print("Shape:", arr.shape)
print("Number of dimensions:", arr.ndim)
print("Data type:", arr.dtype)
print("Size:", arr.size)

arr = np.array([[1,2,3],
                [4,5,True]])

print("Shape:", arr.shape)
print("Number of dimensions:", arr.ndim)
print("Data type:", arr.dtype)
print("Size:", arr.size)

arr = np.array([[1,2,3],
                [4,5,"6"]])

print("Shape:", arr.shape)
print("Number of dimensions:", arr.ndim)
print("Data type:", arr.dtype)
print("Size:", arr.size)

arr = np.array([[1,2,3],
                [4,5,6.0]])

print("Shape:", arr.shape)
print("Number of dimensions:", arr.ndim)
print("Data type:", arr.dtype)
print("Size:", arr.size)
```

Array Reshaping

```python
# arr = np.array([[1,2,3],
#                 [4,5,6.0]], dtype=int)  

# print("Shape:", arr.shape)
# print("Number of dimensions:", arr.ndim)    
# print("Data type:", arr.dtype)
# print("Size:", arr.size)

arr = np.arange(12)
print("Original array:", arr)   
arr_reshaped = arr.reshape((3, 4))
print("Reshaped array (3, 4):\n", arr_reshaped)

flattened = arr_reshaped.flatten()
print("Flattened array:", flattened)

# ravel (returns view, not a copy)
reveled = arr_reshaped.ravel()
print("Raveled array:", reveled)

# Transpose
transposed = arr_reshaped.T
print("Transposed array:\n", transposed)

# adjoint (conjugate transpose)
adjoint = arr_reshaped.conj().T
print("Adjoint array:\n", adjoint)
```
